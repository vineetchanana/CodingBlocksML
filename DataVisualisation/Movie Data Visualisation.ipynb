```python
titles = list(titles)
```

```python
freq_titles = {}
# for title in titles:
    
#     length = len(title)
#     if freq_titles.get(length) is None:
#         freq_titles[length] = 1
#     else:
#         freq_titles[length] +=1
    
for title in titles:
    length = len(title)
    freq_titles[length] = freq_titles.get(length,0) + 1
    
freq_titles
```

```python
length = np.array(list(freq_titles.keys()))
print(length)
```

```python
frequency = np.array(list(freq_titles.values()))

print(frequency)
```

```python
plt.style.use('seaborn')
plt.figure(figsize=(9,7))
plt.xlabel('Length of movie title')
plt.ylabel('Freq/Count of movies')
plt.title('Movie Data Visualisation')
plt.scatter(length,frequency,color='red',edgecolor='black')
plt.show()
```
