```python
x = np.arange(20)
y = x**2
```

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
xpos = np.arange(len(company))
xpos
```

```python
plt.bar(company,revenue,label='Revenue')

plt.legend()
plt.show()
```

```python
plt.bar(xpos,revenue,label='Revenue')
plt.xticks(xpos,company)
plt.legend()
plt.show()
```

```python
profit = [110,70,90,50]
plt.bar(company,revenue,label='revenue')
plt.bar(company,profit,label='profit')

plt.legend()
```

```python
#plt.xticks(xpos,company)
plt.bar(xpos-0.2,revenue,width=0.4,label='revenue')
plt.bar(xpos+0.2,profit,width=0.4,label='profit')

plt.legend()
```

```python
employees=[61,71,79,91,93,89,90,94,99,128,118,114,124,131]
year=[2005,2006,2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018]
revenue=[39.79,44.28,51.12,60.42,58.44,62.48,69.94,73.72,77.85,86.83,93.58,85.32,89.95,110.36]
```

```python
colors = np.arange(len(year))
plt.scatter(year,revenue,s=employees,c=colors,edgecolor='black',alpha=0.6)
plt.xlabel('Year')
plt.ylabel('Revenue')
plt.grid()
for i in range(len(year)):
    plt.text(year[i]+0.1,revenue[i]+0.6,employees[i])
plt.show()
```

```python
labels = ['A','B','C','D']
sizes = [3,4,6,2]
plt.pie(sizes)
#to equal the axes
plt.axis('equal')
plt.show()
```
