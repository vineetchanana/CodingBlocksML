```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Xsn = np.random.randn(100)
u = 60
sigma = 5
X = np.round(Xsn*sigma + u)

X2 = np.round(Xsn*8 + 40)
print(X)
```

```python
plt.style.use('seaborn')
plt.figure(figsize=(10,7))
plt.hist(X,edgecolor='black',alpha=0.8,label='Maths')
plt.hist(X2,edgecolor='black',alpha=0.5,label='Physics')
plt.legend()
plt.show()
```
