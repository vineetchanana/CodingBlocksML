```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
x = np.arange(10)
y1 = x**2
y2 = 2*x+3
print(x)
print(y1)
print(y2)
```

```python
themes = plt.style.available
print(themes)
```

## Line Plot

```python
#use plt.show because it wil finish plotting 1st graph, helps to iterate over to next graph 
plt.plot(x,y1)
plt.show()

plt.plot(x,y2)
plt.show()
```

```python
plt.style.use('seaborn')
```

```python
plt.plot(x,y1,color='red',label='Apple',marker='o')
plt.plot(x,y2,color='green',label='Kiwi',linestyle='dashed')
plt.xlabel('Time')
plt.ylabel('Price')
plt.title('Price of fruits over time')
plt.legend()
plt.show()
```

## Scatter Plot

```python
plt.scatter(x,y1,color='red',label='Apple',marker='^')
plt.scatter(x,y2,color='green',label='Kiwi',marker='+',alpha=0.7)
plt.xlabel('Time')
plt.ylabel('Price')
plt.title('Price of fruits over time')
plt.legend()
plt.show()
```

#### Change the size of any plot

```python
plt.figure(figsize=(10,5))
plt.plot(x,y1,color='red',label='Apple',marker='o')
plt.plot(x,y2,color='green',label='Kiwi',linestyle='dashed')
plt.xlabel('Time')
plt.ylabel('Price')
plt.title('Price of fruits over time')
plt.legend()
plt.show()
```

## Bar Graphs

```python
x_coordinates = np.array([0,1,2])*2
plt.bar(x_coordinates,[20,10,15],width=0.5,label='Current Year',tick_label=['Gold','Platinum','Silver'])
plt.bar(x_coordinates+0.5,[10,20,30],width=0.5,label='Next year')

plt.xlabel('Metal')
plt.ylabel('Price')
plt.title('Metal Price Comparison')
plt.legend()
plt.show()
```

```python
plt.style.use('dark_background')
x_coordinates = np.array([0,1,2])*2
plt.bar(x_coordinates,[20,10,15],width=0.5,label='Current Year',tick_label=['Gold','Platinum','Silver'],color='green')
plt.bar(x_coordinates+0.5,[10,20,30],width=0.5,label='Next year',color='red')

#specifying the limit of the graph
plt.ylim(0,50)
plt.xlim(-2,6)

plt.xlabel('Metal')
plt.ylabel('Price')
plt.title('Metal Price Comparison')
plt.legend()
plt.show()
```

## Normal Distribution

```python
Xsn = np.random.randn(100)
#print(Xsn)
sigma = 5
u = 60 #mean
X = np.round(Xsn*sigma + u)
print(X)

X2 = np.round(Xsn*8 + 40)
```

```python
plt.style.use('seaborn')
plt.hist(X,edgecolor='black',alpha=0.8,label='Maths')
plt.hist(X2,edgecolor='black',alpha=0.7,label='Physics')
plt.xlabel('Marks obtained by students')
plt.ylabel('Probab/FreqCount of students')
plt.title('Histogram')
plt.legend()
plt.show()
```
