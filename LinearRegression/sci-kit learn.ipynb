```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import make_regression
```

```python
#to enter interactive mode
%matplotlib notebook
```

```python
X,y = make_regression()

print(X.shape,y.shape)
```

- By default it gives 100 examples and features

```python
#n_informative =1 means only 1 of the feature is not useful
#random_state helps to get same data every time
X,y = make_regression(n_samples=1000,n_features=2,n_informative=2,noise=10,random_state=1)

print(X.shape,y.shape)
```

```python
plt.subplot(1,2,1)
plt.scatter(X[:,0],y)

plt.subplot(1,2,2)
plt.scatter(X[:,1],y)
```

- we can see only first feature is showing some linear relation, while in other feature if we increase x then y have both increasing as well as decreasing values
- as we have  features we can plot the data in 3d

```python
from mpl_toolkits import mplot3d
fig = plt.figure(figsize=(9,6))
ax = plt.axes(projection='3d')

#Create 3d plot
ax.scatter3D(X[:,0],X[:,1],y,color='green')
plt.title('Scatter plot 3d')
plt.show()
```

- from a different angle we can see it's actually a regression data and we can pass a plane through it.
- In this our goal is to find the hyper plane which fits our data
- hypothesis- t0 + t1x1 + t2x2 = y..and this is the equation of the plane

# Implementing Linear regresion using Sci-kit Learn

```python
from sklearn.linear_model import LinearRegression
```

```python
#make an object
model = LinearRegression()
```

```python
#performing gradient descent(training our model)
model.fit(X,y)
```

```python
model.coef_
```

```python
model.intercept_
```

- model.coef_ contains t1,t2,....,tn
- model.intercept_ contains.. t0

```python
model.predict([X[0],X[1]])
```

```python
y[0],y[1]
```

```python
#check the performance on training data since we donot have test data
model.score(X,y)
```
