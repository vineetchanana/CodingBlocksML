```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X = np.arange(20)
```

```python
#gives us points which are normally distributed, by default it gives us points with mean = 0 and variance =1
#for some other mean and variance, do sigma*np.random.randn(20) + mean(u)
noise = 3*np.random.randn(20)
```

```python
plt.style.use('seaborn')
plt.scatter(X,noise)
plt.show()
```

- We can see most of the points are centered around line x =0 (because of zero mean) and are in the range -3 to 3 (because 3 is the variance)

```python
theta = np.array([2,3])
y_ideal = theta[0] + X*theta[1]
y_real = y_ideal + noise
```

```python
plt.plot(X,y_ideal,color='green')
plt.scatter(X,y_real)
plt.show()
```

- Green line is the ideal line(where points lie on the line) but due to some noise, some real life unmodelled effects points now line about the line with some standard deviation
- First these points were centered around y = 0 but we added a term theta'X to every point which shifted the point 
- The points are centered around new mean, mean value is changing as x is changing, if x is large, mean is large
- As we go away from the line in a vertical direction, probability of occuring a point decreases
- We shifted the mean by a value of theta'X

```python
plt.plot(X,y_ideal,color='green')
plt.scatter(X,y_real)
plt.scatter(X,noise,color='red')  # red shows original error points
plt.show()
```
