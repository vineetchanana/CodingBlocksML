Surface plots are used

- to visulaize Loss function in Machine learning and deep learning
- to visualize state and state value functions in Reinforcement learning

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
a = np.array([1,2,3])
b = np.array([4,5,6,7])

a,b = np.meshgrid(a,b)

print(a)
print(b)
```

Both a and b have dimensions 4*3, 'a' got repeated along rows 'b' number of times and 'b' got broadcasted along columns 'a' number of times

We can use both matrices to plot a figure in 3d

```python
#give us axes in 3d
fig = plt.figure()
axes = fig.gca(projection='3d')
#axes.plot_surface(a,b,a+b)
plt.show()
```

```python
fig = plt.figure()
axes = fig.gca(projection='3d')
axes.plot_surface(a,b,a+b,cmap='coolwarm')
plt.show()
```

red means highest value and blue means lowest value

```python
fig = plt.figure()
axes = fig.gca(projection='3d')
axes.plot_surface(a,b,a+b,cmap='rainbow')
plt.show()
```

```python
#earlier it was linear, now it is not linear 
fig = plt.figure()
axes = fig.gca(projection='3d')
axes.plot_surface(a,b,a**2+b**2,cmap='rainbow')
plt.show()
```

```python
x = np.arange(-1,1,0.02)
y = x

x,y = np.meshgrid(x,y)
print(x.shape,y.shape)
```

```python
fig = plt.figure()
axes = fig.gca(projection='3d')
axes.plot_surface(x,y,x**2+y**2,cmap='rainbow')
plt.show()
```

minima is on 0,0 and symmetric around 0,0 (as square negaive sign doesnot matter)when we go towards 1 or -1 allegation increases

## Contour plots

In contour plots we cut this by parallel planes and same values of z will be denoted by same color

```python
fig = plt.figure()
axes = fig.gca(projection='3d')
axes.contour(x,y,x**2+y**2,cmap='rainbow')
plt.show()
```
