```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.arange(10)
Y = (X-5)**2
```

```python
print(X)
print(Y)
```

```python
plt.style.use('seaborn')
plt.xlabel('X')
plt.ylabel('Y')
plt.plot(X,Y)
plt.show()
```

```python
x = 0
lr = 0.1
error = []
plt.plot(X,Y)
for i in range(50):
    grad = 2*(x-5)
    x = x - lr*grad
    y = (x-5)**2
    error.append(y)
    plt.scatter(x,y)
    #print(x)
```

## Plot values of error

```python
plt.plot(error)
plt.show()
```

After 20 iterations, the value of y is almost constant,we can put check to stop the gradient descent early
