```python
def snapchat(image):
    image_copy = image.copy()
    sunglasses_copy = sunglasses.copy()
    mustache_copy = mustache2.copy()
    face_rects = face_cascade.detectMultiScale(image_copy,1.3,5)

    for (x,y,w,h) in face_rects:
        image_face = image_copy[y:y+h,x:x+w]

    plt.imshow(image_face)
    plt.show()
    eyes = eyes_cascade.detectMultiScale(image_face,1.1,10)
    nose = nose_cascade.detectMultiScale(image_face,1.3,10)


    for eye in eyes:
        ex,ey,ew,eh = eye
        eye_area = w*h
        print(eye,ew,eh)

        print(sunglasses_copy.shape)
        sunglasses_copy = cv2.resize(sunglasses_copy,(ew,eh))
        print(sunglasses_copy.shape)
        print(jamie_face.shape)
        #cv2.rectangle(jamie_face,(ex,ey),(ex+ew,ey+eh),(255,0,0),2)



        for i in range(sunglasses_copy.shape[0]):
            for j in range(sunglasses_copy.shape[1]):
                if sunglasses_copy[i,j,3] > 0:

                    image_face[ey+i,ex+j,:] = sunglasses_copy[i,j,:-1]

    for (nx,ny,nw,nh) in nose:
        hh = nh/2
        print(nh,hh)
        mustache_copy = cv2.resize(mustache_copy,(nw,nh))

        for i in range(mustache_copy.shape[0]):
            for j in range(mustache_copy.shape[1]):
                if mustache_copy[i,j,3] > 0:
                    image_face[i+ny+int(nh/2),j+nx,:] = mustache_copy[i,j,:-1]



    
    plt.imshow(image_copy)
    plt.show()
    return image_copy
```
