- we need to calculate conditional probability for the features and we can do that by using mormal distribution formula (we need sigma and mean value)

```python
from sklearn.datasets import make_classification
X,Y = make_classification(n_samples=200,n_features=2,n_informative=2,n_redundant=0,random_state=3)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(X[:,0],X[:,1],c=Y)
plt.show()
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gnb = GaussianNB()
```

```python
gnb.fit(X,Y)
```

```python
Y_pred = gnb.predict(X)
Y_pred
```

```python
gnb.score(X,Y)
```

- we are getting 94% accuracy, score method internally calculates y_mpred and compares with the given Y

```python
import numpy as np
acc = np.sum(Y_pred==Y)/X.shape[0]
acc
```

## Generate Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
```

```python
cnf_matrix = confusion_matrix(Y,Y_pred)
cnf_matrix
```
