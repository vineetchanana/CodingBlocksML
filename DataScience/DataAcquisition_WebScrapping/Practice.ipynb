```python
from urllib.request import urlopen
```

```python
vampire_url = 'https://en.wikipedia.org/wiki/List_of_vampire_films'
```

```python
vampire_data = urlopen(vampire_url)
```

```python
vampire_data
```

```python
vampire_html = vampire_data.read()
vampire_html
```

```python
vampire_data.close()
```

```python
from bs4 import BeautifulSoup as soup
```

```python
vampire_soup = soup(vampire_html,'html.parser')
```

```python
tables = vampire_soup.findAll('table',{'class':'wikitable sortable'})
```

```python
len(tables)
```

```python
vampire_table = tables[0]
```

```python
headers = vampire_table.findAll('th')
headers
```

```python
headers[0].text[:-1]
```

```python
column_titles = [ct.text[:-1] for ct in headers]
column_titles
```

```python
rows = vampire_table.findAll('tr')
len(rows)
```

```python
first_row = rows[1]
first_row
```

```python
table_rows = []

for row in rows[2:]:
    current_row = []
    row_data = row.findAll('td')
    for idx,data in enumerate(row_data):
        
        current_row.append(data.text)
        
    table_rows.append(current_row)
        
```

```python
#create csv
```

```python
with open('vampire.txt','w') as f:
    header_string = ','.join(column_titles)
    header_string += '\n'
    f.write(header_string)
    
    for r in rows:
        row_string = ''
        row_string = ','.join(r)
        row_string += '\n'
        f.write(row_string)
```
