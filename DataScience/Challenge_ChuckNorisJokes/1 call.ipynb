Previously, I was making call for each joke(was taking so much of time), now i am extracting all jokes at once then filter out the required jokes

```python
import requests
```

```python
import json
```

```python
import pandas as pd
```

```python
url = 'http://api.icndb.com/jokes'
```

```python
response = requests.get(url)
```

```python
json_data = json.loads(response.content)
```

```python
json_data['value']
```

```python
json_data = json_data['value']
```

```python
json_data[0]
```

```python
id_dataframe = pd.read_csv('ID.csv')
```

```python
id_dataframe.head()
```

```python
id_dataframe.shape[0]
```

```python
jokes = []
i=0
for idx,data in enumerate(json_data):
    #print(data['id'])
    if i >= id_dataframe.shape[0]:
        break
    if data['id'] == id_dataframe['ID'][i]:
        #print(data['joke'])
        jokes.append((data['id'],data['joke']))
        i +=1
```

```python
jokes[10]
```

```python
len(jokes)
```

```python
column_titles = ['ID','Joke']
```

```python
with open('jokes_better.csv','w') as file:
    #write headers
    header_string = ','.join(column_titles) + '\n'
    file.write(header_string)
    
    for index,joke in jokes:
        joke_string = str(index) + ','
        if ',' in joke:
            joke_string += '"' + joke + '"'
            joke_string += '\n'
        else:
            joke_string += joke + '\n'
        #print(joke_string)
        file.write(joke_string)
```
