API - set of routines running over web server

API provides us to get data like facebook,twitter

We make a API call from web browser and API interacts with database and gives back us the data

Here we will make API calls from python script

2 ways to make url requests

- urllib.requests library
- requests library(famous)

```python
from urllib.request import urlopen
```

```python
api_url = 'https://samples.openweathermap.org/data/2.5/weather?q=London,uk&appid=439d4b804bc8187953eb36d2a8c26a02'
```

```python
url_result = urlopen(api_url)
data = url_result.read()
```

```python
print(url_result)
```

```python
print(data)
```

```python
print(type(data))
```

```python
#convert this into json(python dictionary)
```

```python
import json
```

```python
json_data = json.loads(data)
print(json_data)
```

```python
json_data['coord']  #dictionary
```

```python
json_data['name']
```

if you want to convert json data to a string(to save it into a file)

```python
json_string = json.dumps(json_data)
print(json_string)
```

```python
type(json_string)
```

```python
print(r.content)
```

# Google API

```python
url = 'https://maps.googleapis.com/maps/api/geocode/json?'
```

```python
#requests allow us to pass dictionary as parameters
```

```python
parameters = {
    'address' : 'Coding Blocks Pitampura',
    'key' : 'iDontHaveAny'
}
```

```python
r = requests.get(url,params=parameters)
```

```python
r.url
```

```python
r.content
```

```python
r.content.encode('UTF-8')
```
